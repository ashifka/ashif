### **Check and handle outliers in at least 3 columns in the dataset**

```python
#writing any first 2 columns in Z-Score Method of handling Outliers
column1 =imputed_value['Age']
z_scores1 = (column1 - column1.mean()) / column1.std()
z_scores1.mean()
```

```python
#putting threshold as 3
threshold = 3
outliers1 = imputed_value[abs(z_scores1) > threshold]
```

```python
print("Outliers of column 'Age' is:")
outliers1
```

```python
column2 =imputed_value['Fare']
z_scores2 = (column2 - column2.mean()) / column2.std()
z_scores2.mean()
```

```python
outliers2 = imputed_value[abs(z_scores2) > threshold]
print("Outliers of column 'Fare' is:")
outliers2
```

```python
#writing third column in IQR Method of handling Outliers
column3 = imputed_value['SibSp']
Q1 = column3.quantile(0.25)
Q3 = column3.quantile(0.75)
IQR = Q3 - Q1
```

```python
outliers3 = imputed_value[(column3 < Q1 - threshold * IQR) | (column3> Q3 + threshold * IQR)]
print("Outliers of column 'SibSp' is:")
outliers3
```

### **Do min max scaling on the feature set (Take ‘Survived’ as target)**

```python
#first we have to remove string columns
columns_to_remove = ['Name', 'Sex', 'Ticket','Cabin','Embarked']
imputed_value.drop(columns=columns_to_remove, axis=1, inplace=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
target = imputed_value['Survived']
features = imputed_value.drop('Survived',axis =1)
scaler = MinMaxScaler()
```

```python
scaled_features = scaler.fit_transform(features)
scaled_normalized = pd.DataFrame(scaled_features, columns=features.columns)
final_normalized = pd.concat([scaled_normalized, target], axis=1)
```

```python
final_normalized
```

```python
final_normalized.fillna(final_normalized.mean(),inplace = True)
```
