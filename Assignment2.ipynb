# **2.**
 A company started to invest in digital marketing as a new
way of their product promotions. For that they collected
data and decided to carry out a study on it.
# **a.**
The company wishes to clarify whether there is any
increase in sales after stepping into digital marketing.



```python
from scipy.stats import ttest_rel
```

```python
data2 = pd.read_csv('/content/sample_data/Sales_add.csv')
```

```python
data2
```

```python
t_stats,p_value = ttest_rel(data2['Sales_After_digital_add(in $)'],data2['Sales_before_digital_add(in $)'])
```

```python
p_value
```

```python
if p_value<0.05:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```

# **b.**
 The company needs to check whether there is any
dependency between the features “Region” and
“Manager”.

```python
from scipy.stats import chi2_contingency
```

```python
chi_table = pd.crosstab(data2['Region'],data2['Manager'])
```

```python
chi_table
```

```python
chi2_stat,p_value1,dof,exp = chi2_contingency(chi_table)
```

```python
p_value1
```

```python
if p_value1<0.05:
  print('Reject Null Hypothesis')
else:
  print('Accept Null Hypothesis')
```
